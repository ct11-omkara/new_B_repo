```python
dots_data = pd.read_excel("Data//September_09//KapanGrading_dots.xlsx")
inclusion_data = pd.read_excel("Data//September_09//KapanGrading_inclusion.xlsx")
LabBreakage_data = pd.read_excel("Data//September_09//KapanGrading_LabBreakage.xlsx")
```

```python
dots_data.fillna(0, inplace=True)
inclusion_data.fillna(0, inplace=True)
LabBreakage_data.fillna(0, inplace=True)
```

```python
dots_data.dots_heavy_prct = pd.to_numeric(dots_data.dots_heavy_prct)
dots_data.dots_medium_prct = pd.to_numeric(dots_data.dots_medium_prct)
dots_data["dots_total"] = dots_data["dots_heavy_prct"] + dots_data["dots_medium_prct"] 
```

```python
inclusion_data.inclusion_heavy_prct = pd.to_numeric(inclusion_data.inclusion_heavy_prct)
inclusion_data.inclusion_medium_prct = pd.to_numeric(inclusion_data.inclusion_medium_prct)
inclusion_data["inclusion_total"] = inclusion_data["inclusion_heavy_prct"] + inclusion_data["inclusion_medium_prct"] 
```

```python
LabBreakage_data.LabBreakage_heavy_prct = pd.to_numeric(LabBreakage_data.LabBreakage_heavy_prct)
LabBreakage_data.LabBreakage_medium_prct = pd.to_numeric(LabBreakage_data.LabBreakage_medium_prct)
LabBreakage_data["labBreakage_total"] = LabBreakage_data["LabBreakage_heavy_prct"] + LabBreakage_data["LabBreakage_medium_prct"] 
```

```python
final_quality = pd.merge(pd.merge(dots_data,inclusion_data,how="left",on=["Machine"]),LabBreakage_data,how="left",on=["Machine"])
```

```python
def add_sorting_rank_to_column(temp_df, column_name, ascending_order=True):
    temp_df[column_name] = pd.to_numeric(temp_df[column_name])
    temp_df = temp_df.sort_values(by=[column_name], ascending=ascending_order).reset_index()
    temp_df.index.name = column_name + "_rank"
    temp_df = temp_df.reset_index()
    temp_df[column_name + "_rank"] += 1
    temp_df.index.name = "delete_" + column_name
    return temp_df
```

```python
final_quality.dots_total = pd.to_numeric(final_quality.dots_total)
final_quality.inclusion_total = pd.to_numeric(final_quality.inclusion_total)
final_quality.labBreakage_total = pd.to_numeric(final_quality.labBreakage_total)
```

```python
final_quality = add_sorting_rank_to_column(final_quality,"dots_total")
final_quality = add_sorting_rank_to_column(final_quality,"inclusion_total")
final_quality = add_sorting_rank_to_column(final_quality,"labBreakage_total")
```
