```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
quantity = pd.read_excel("quantityReport_for_only_machine.xlsx", index_col="ID")
zone = pd.read_excel("zoneReport.xlsx", index_col="ID")
quality = pd.read_excel("qualityReport.xlsx", index_col="ID")
color = pd.read_excel("colorReport_2.xlsx", index_col="ID")
```

```python
final_merging = pd.merge(pd.merge(quantity,quality,how="left",on=["Machine"]),
                         pd.merge(color,zone,how="left",on=["Machine"]),how="left",on=["Machine"])
```

```python
final_merging[final_merging.isna().any(axis=1)]
```

```python
if final_merging.index[0]==0:
    final_merging.index += 1
```

```python
final_merging
```

```python
final_merging.GrRate_rank = pd.to_numeric(final_merging.GrRate_rank)
final_merging.GrownWeight_sum_rank = pd.to_numeric(final_merging.GrownWeight_sum_rank)
final_merging.Actual_Hours_avg_rank = pd.to_numeric(final_merging.Actual_Hours_avg_rank)
final_merging.GrowingHours_sum_rank = pd.to_numeric(final_merging.GrowingHours_sum_rank)
final_merging.GrownHeight_sum_rank = pd.to_numeric(final_merging.GrownHeight_sum_rank)
final_merging.Dmtr_avg_rank = pd.to_numeric(final_merging.Dmtr_avg_rank)
final_merging.dots_total_rank = pd.to_numeric(final_merging.dots_total_rank)
final_merging.inclusion_total_rank = pd.to_numeric(final_merging.inclusion_total_rank)
final_merging.labBreakage_total_rank = pd.to_numeric(final_merging.labBreakage_total_rank)
final_merging.Zone_total_rank = pd.to_numeric(final_merging.Zone_total_rank)
```

```python
final_merging['Final_total_rank'] = final_merging['GrRate_rank'] + final_merging['GrownWeight_sum_rank'] + final_merging['Actual_Hours_avg_rank'] + \
final_merging['GrowingHours_sum_rank'] + final_merging['GrownHeight_sum_rank'] + final_merging['Dmtr_avg_rank'] + final_merging['dots_total_rank'] + \
final_merging['inclusion_total_rank'] + final_merging['labBreakage_total_rank'] + final_merging['Zone_total_rank']
```
