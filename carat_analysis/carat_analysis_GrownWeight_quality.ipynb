```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from collections import Counter
import os
```

```python
mainDF = pd.read_excel("data/November_11/LabHoursData.xlsx")
dots = pd.read_excel("data/November_11/KapanGradinPivot_dots.xlsx")
inclusions = pd.read_excel("data/November_11/KapanGradinPivot_inclusions.xlsx")
labBreakages = pd.read_excel("data/November_11/KapanGradinPivot_LabBreakages.xlsx")
mainDF.columns
```

```python
dots.fillna(0, inplace=True)
inclusions.fillna(0, inplace=True)
labBreakages.fillna(0, inplace=True)
```

```python
labBreakages.isnull().sum()
```

```python
mainDF
```

```python
#remove = "BreakdownTime", "GrowingHours"
final_DF = mainDF[["Machine", "GrownWeight", "Dmtr", "GrowingHours", "GrownHeight", "FinalWeight", "IniAvgHeight",
                    "BreakdownTime", "IniWeight", "Width", "Length", "Actual Hours", "GrRate"]]
#final_DF[final_DF.isna().any(axis=1)]
final_DF = final_DF.rename(columns = {'Actual Hours':'Actual_Hours'})

final_DF.tail(2)
```

```python
nullID = final_DF[final_DF[['Machine','GrownWeight', 'GrownHeight','Dmtr', \
                            'Actual_Hours', 'GrRate']].isna().any(axis=1)].index
nullID
```

```python
final_DF = final_DF.drop(nullID)
```

```python
final_DF.tail(2)
```

```python
final_DF.GrownWeight = pd.to_numeric(final_DF.GrownWeight)
final_DF.GrowingHours = pd.to_numeric(final_DF.GrowingHours)
final_DF.FinalWeight = pd.to_numeric(final_DF.FinalWeight)
final_DF.ProdHeight = pd.to_numeric(final_DF.GrownHeight)
final_DF.Dmtr = pd.to_numeric(final_DF.Dmtr)
final_DF.Actual_Hours = pd.to_numeric(final_DF.Actual_Hours)
final_DF.GrRate = pd.to_numeric(final_DF.GrRate)
final_DF.IniAvgHeight = pd.to_numeric(final_DF.IniAvgHeight)
final_DF.BreakdownTime	 = pd.to_numeric(final_DF.BreakdownTime	)
final_DF.IniWeight = pd.to_numeric(final_DF.IniWeight)
final_DF.Width = pd.to_numeric(final_DF.Width)
final_DF.Length = pd.to_numeric(final_DF.Length)
```

```python
dots['dots_heavy_prct'] = pd.to_numeric(dots['dots_heavy_prct'])
dots['dots_medium_prct'] = pd.to_numeric(dots['dots_medium_prct'])
dots['dots_minor_prct'] = pd.to_numeric(dots['dots_minor_prct'])

inclusions['inclusions_heavy_prct'] = pd.to_numeric(inclusions['inclusions_heavy_prct'])
inclusions['inclusions_medium_prct'] = pd.to_numeric(inclusions['inclusions_medium_prct'])
inclusions['inclusions_minor_prct'] = pd.to_numeric(inclusions['inclusions_minor_prct'])

labBreakages['lab_breakages_heavy_prct'] = pd.to_numeric(labBreakages['lab_breakages_heavy_prct'])
labBreakages['lab_breakages_medium_prct'] = pd.to_numeric(labBreakages['lab_breakages_medium_prct'])
labBreakages['lab_breakages_minor_prct'] = pd.to_numeric(labBreakages['lab_breakages_minor_prct'])
```

```python
df_quality = pd.merge(pd.merge(
        dots.groupby(['Machine'])['dots_heavy_prct','dots_medium_prct','dots_minor_prct'].mean().reset_index(),
        inclusions.groupby(['Machine'])['inclusions_heavy_prct', 'inclusions_medium_prct', 'inclusions_minor_prct'].mean().reset_index(), 
                                      how="left", on=["Machine"]),
        labBreakages.groupby(['Machine'])['lab_breakages_heavy_prct','lab_breakages_medium_prct','lab_breakages_minor_prct'].mean().reset_index(),
                                     how="left", on=["Machine"])
```

```python
df_machine = pd.merge(
        final_DF.groupby(['Machine'])['GrownWeight','GrowingHours','GrownHeight','FinalWeight'].sum().reset_index(),
        final_DF.groupby(['Machine'])['GrRate', 'IniAvgHeight', 'BreakdownTime', 'Actual_Hours', 'IniWeight', 'Width', 'Length','Dmtr'].mean().reset_index(), 
                                      how="left", on=["Machine"])
```

```python
df_machine = pd.merge(df_machine, df_quality, how="left", on=["Machine"])
df_machine
```

```python
#65% of grown weight (means 35% less)
df_machine["GrownWeight_less_35_prct"] = df_machine["GrownWeight"] * 65 / 100
```

```python
df_machine
```

```python
final = df_machine[(df_machine.GrownWeight_less_35_prct>=200) & (df_machine.GrownWeight_less_35_prct<201)]
final = pd.DataFrame(final.describe().T)[["mean"]].rename(columns={"mean":"Average"})
if os.path.exists("Equal_to_200.xlsx"):
    os.remove("Equal_to_200.xlsx")
    
final.to_excel("Equal_to_200.xlsx", header=1, index_label="columns")
```

```python
final = df_machine[df_machine.GrownWeight_less_35_prct<201]
final = pd.DataFrame(final.describe().T)[["mean"]].rename(columns={"mean":"Average"})
if os.path.exists("less_than_200_carats.xlsx"):
    os.remove("less_than_200_carats.xlsx")
    
final.to_excel("less_than_200_carats.xlsx", header=1, index_label="columns")
```
